```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
data = pd.read_csv('real_estate_price_size.csv')
```

```python
data
```

```python
data.describe()
```

```python
y = data['price']
x1 = data['size']
```

```python
plt.scatter(x1,y)
plt.xlabel('size', fontsize = 20)
plt.ylabel('price', fontsize = 20)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat = 223.1787*x1 + 1.019e+05
fig = plt.plot(x1, yhat, lw = 4, c = 'orange', label = 'regression line')
plt.xlabel('size', fontsize = 20)
plt.ylabel('price', fontsize = 20)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('real_estate_price_size.csv')
data.head()
```

```python
y = data['price']
x = data['size']
```

```python
x.shape
```

```python
y.shape
```

```python
x = x.values.reshape(-1,1)
x.shape
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[1000]])
```

```python
new_data = pd.DataFrame(data = [2763,1700], columns = ['size'])
new_data
```

```python
model.predict(new_data)
```

```python
new_data['predicted_price'] = model.predict(new_data)
new_data
```
