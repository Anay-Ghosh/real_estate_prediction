Multicollinearity

```python
data_cleaned = data_cleaned.drop(['Price'],axis=1)
data_cleaned.columns.values
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage','Year','EngineV']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values,i) for i in range(variables.shape[1])]
vif['Features'] = variables.columns
vif
```

```python
data_no_multicollinearity = data_cleaned.drop(['Year'],axis=1)
data_no_multicollinearity
```

Creating dummies

```python
data_with_dummies = pd.get_dummies(data_no_multicollinearity,drop_first=True)
data_with_dummies.head()
```

```python
data_with_dummies.columns.values
```

```python
col = ['log_price','Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[col]
data_preprocessed.head()
```

#Linear Model

Select the target and input

```python
target = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'],axis=1)
```

Scale the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(inputs)
```

```python
input_scaler = scaler.transform(inputs)
```

Split the data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(input_scaler,target,test_size=0.2,random_state=365)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_hat = model.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel('Targets(y_train)',size=18)
plt.ylabel('Predictions(y_hat)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
sns.distplot(y_train-y_hat)
plt.title('Residuals',size=18)
plt.show()
```

```python
model.score(x_train,y_train)
```

Finding Weight and bias

```python
model.coef_
```

```python
model.intercept_
```

```python
model_summary = pd.DataFrame(inputs.columns.values,columns=['Features'])
model_summary['Weights'] = model.coef_
model_summary
```

Testing

```python
y_hat_test = model.predict(x_test)
```

```python
plt.scatter(y_test,y_hat_test,alpha=0.2)
plt.xlabel('Targets(y_test)',size=18)
plt.ylabel('Predictions(y_hat_test)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test),columns=['Predictions'])
df_pf.head()
```

```python
df_pf['Targets'] = np.exp(y_test)
df_pf.head()
```

```python
y_test = y_test.reset_index(drop=True)
y_test.head()
```

```python
df_pf['Targets'] = np.exp(y_test)
df_pf.head()
```

```python
df_pf['Residential'] = df_pf['Targets']-df_pf['Predictions']
```

```python
df_pf['Difference%'] = np.absolute(df_pf['Residential']/df_pf['Targets']*100)
df_pf.head()
```
